```python
#  Importing all nessesory library
import pandas as pd
import numpy as np
import spacy
import string
from spacytextblob.spacytextblob import SpacyTextBlob
nlp = spacy.load('en_core_web_sm')
punct = string.punctuation
spacy_text_blob = SpacyTextBlob()
nlp.add_pipe(spacy_text_blob)

df = pd.read_excel('AllTranscriptions.xlsx')  #reading reviews dataset
print(df.shape)
print("File has read succesfull")
```

```python

# cleaning text
def text_clean(text):
    text = text.lower()  #Convert text in lower case
    punc_removed = [char for char in text if char not in punct]  #Removing Punctuations
    punc_removed_join = ''.join(punc_removed) 
    
    doc= nlp(punc_removed_join)
    text_out = [token.lemma_ for token in doc if token.is_stop == False and token.is_alpha and len(token)>2]
    txt = ' '.join(text_out)
    return txt
# Checking Polarity
def polarity(text):
    doc = nlp(text)
    pol = float(format(doc._.sentiment.polarity, '.3f'))
    return pol

# getting Emotion
def emotion_analysis(text):
    emotion_list = []
    with open('Emotions.txt', 'r') as file:
        for line in file:
            clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
            word, emotion = clear_line.split(':')
        
            if word in text:
                emotion_list.append(emotion)
    return emotion_list


df['clean_text'] = df['Text'].apply(text_clean) #text Cleaning
print("Text Cleaning done successfully")

df['polarity'] = df['clean_text'].apply(polarity)  #polarity checking
df['sentiment'] = df['polarity'].apply(lambda x: "Pos" if x>0 else("Neg" if x<0 else "Neu") )
print("polarity checked successfull")

df['emotions'] = df['clean_text'].apply(emotion_analysis) #getting emotions
df['emotions'] = df['emotions'].apply(lambda x: ', '.join(x))
print("final resulting data")
df.sample(10)
```
